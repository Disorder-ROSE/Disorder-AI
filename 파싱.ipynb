```python
import os
import json
import shutil
from PIL import Image
import xml.etree.ElementTree as ET
```

```python
path = "./Dataset/라벨링데이터/"
file_list = os.listdir(path)
print(file_list)
```

```python
test_list = []
for i in range(len(file_list)):
    path = "./Dataset/라벨링데이터/" + str(file_list[i]) + "/"
    print(path)
    snack_list = os.listdir(path)
    file_list_py = [file for file in snack_list if file.endswith("meta.xml")]
    print ("file_list_py: {}".format(file_list_py))
    test_list.append(file_list_py)
```

```python
for j in range(len(test_list)):
    for k in range(len(test_list[j])):
        tree = ET.parse('./Dataset/라벨링데이터/{}/{}'.format(file_list[j], test_list[j][k]))
        root = tree.getroot()
        sizew = int(root[2][4][0].text)
        sizeh = int(root[2][4][1].text)
        cnt = int(root[2][4][2].text)
#         sizew, sizeh, cnt
        xmin = int(root[2][6][4][0].text)
        ymin = int(root[2][6][4][1].text)
        xmax = int(root[2][6][4][2].text)
        ymax = int(root[2][6][4][3].text)
#         xmin, xmax, ymin, ymax
        xcen = (xmin+xmax) / 2 / sizew
        ycen = (ymin+ymax) / 2 / sizeh
#         xcen, ycen
        w = (xmax-xmin) / sizew
        h = (ymax-ymin) / sizeh
#         w, h
        line = str(j), ' ', str(xcen), ' ', str(ycen), ' ', str(w), ' ', str(h)
        line = ''.join(line)
        with open("./Dataset/label/{}/{}.txt".format(file_list[j], test_list[j][k][:-9]), "w") as file:
            file.write(line)
            file.close()
```
