```python
import cv2
import os
```

```python
def resizee(imgpath, name):
    img = cv2.imread(imgpath + name[:-4]+ ".jpg")
    
    width = 640
    height = 640

    resized_image = cv2.resize(img, (width, height))

    cv2.imwrite(imgpath + name[:-4] + ".jpg", resized_image)
```

```python
path = "./Dataset/Dataset/valid/images"
flist = os.listdir(path)
print(flist)
```

```python
for k in range(len(flist)):
    imgpath = "./Dataset/Dataset/valid/images/" + flist[k] + "/"
    file_list = os.listdir(imgpath)
    print(file_list)
    for i in range(len(file_list)):
        resizee(imgpath, file_list[i])
```
